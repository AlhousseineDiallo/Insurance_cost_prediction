## let's start the training of the model

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=23
)
```

```python
log_reg = LogisticRegression(solver="liblinear")
```

```python
log_reg.fit(X_train, y_train)
```

```python
# let's test the accuracy of the model
log_reg.predict_proba(X_train)
```

```python
y_pred_train = log_reg.predict(X_train)
```

```python
# let's test the accuracy of the model
log_reg.score(X_train, y_train)
```

```python
# let's test the accuracy score on the test set
log_reg.score(X_test, y_test)
```

```python
# let's use the accuracy method to do this also
# on the train set first
print(accuracy_score(y_train, y_pred_train))
```

```python
y_pred_test = log_reg.predict(X_test)
```

```python
# on the test set
print(accuracy_score(y_test, y_pred_test))
```
